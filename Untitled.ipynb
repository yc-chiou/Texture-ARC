```python
# ARC Simulation

from __future__ import division

import argparse
import meep as mp
import time
import numpy as np
import matplotlib.pyplot as plt
from meep.materials import  cSi, SiO2
```

```python
sx=10
sy=10
eps_Si=12
eps=4
T_Si=1
T_Arc=0.1
cell = mp.Vector3(sx, sy, 0)
pml_layers = [mp.PML(1.0)]
resolution = 50
nfreq = 500
fcen = 1.75  # pulse center frequency
df = 1.5  # pulse width (in frequency)
```

```python
#geometry = [mp.Block(mp.Vector3(T_Si, sy, mp.inf), center=mp.Vector3(0, 0), material=mp.Medium(epsilon=eps_Si))]
geometry = [mp.Block(mp.Vector3(T_Si, sy, mp.inf), center=mp.Vector3(0, 0), material=cSi)]
```

```python
sources = [mp.Source(mp.GaussianSource(fcen, fwidth=df), component=mp.Ez, 
                     center=mp.Vector3(2 + (-0.5 * sx), 0), size=mp.Vector3(0, sy))]
```

```python
sim = mp.Simulation(cell_size=cell,
                    boundary_layers=pml_layers,
                    geometry=geometry,
                    sources=sources,
                    resolution=resolution)
refl_1_fr = mp.FluxRegion(center=mp.Vector3(-0.51, 0), size=mp.Vector3(0, sy))
refl_1 = sim.add_flux(fcen, df, nfreq, refl_1_fr)
tran_1_fr = mp.FluxRegion(center=mp.Vector3(0, 0), size=mp.Vector3(0, sy))
tran_1 = sim.add_flux(fcen, df, nfreq, tran_1_fr)
plt.figure(dpi=600)
sim.plot2D()
plt.show()
```

```python
pt = mp.Vector3(0, 0)
sim.run(mp.at_beginning(mp.output_epsilon),
                until_after_sources=mp.stop_when_fields_decayed(50, mp.Ez, pt, 1e-3))
no_ARC_refl_data = sim.get_flux_data(refl_1)
```

```python
no_ARC_tran_flux = mp.get_fluxes(tran_1)
no_ARC_refl_flux = mp.get_fluxes(refl_1)
```

```python
sim.reset_meep()
geometry = [mp.Block(mp.Vector3(T_Si, sy, mp.inf), center=mp.Vector3(0, 0),
                            material=cSi),
                            mp.Block(mp.Vector3(T_Arc, sy, mp.inf), center=mp.Vector3(-0.5 * (T_Si + T_Arc), 0),
                            material=SiO2)]
sim = mp.Simulation(cell_size=cell,
                    boundary_layers=pml_layers,
                    geometry=geometry,
                    sources=sources,
                    resolution=resolution)
refl_1_fr = mp.FluxRegion(center=mp.Vector3(-0.61, 0), size=mp.Vector3(0, sy))
refl_1 = sim.add_flux(fcen, df, nfreq, refl_1_fr)
tran_1_fr = mp.FluxRegion(center=mp.Vector3(0, 0), size=mp.Vector3(0, sy))
tran_1 = sim.add_flux(fcen, df, nfreq, tran_1_fr)

sim.load_minus_flux_data(refl_1, no_ARC_refl_data)
pt = mp.Vector3(0, 0)
sim.run(mp.at_beginning(mp.output_epsilon),
                until_after_sources=mp.stop_when_fields_decayed(50, mp.Ez, pt, 1e-3))

with_ARC_refl_flux = mp.get_fluxes(refl_1)
with_ARC_tran_flux = mp.get_fluxes(tran_1)

flux_freqs = mp.get_flux_freqs(refl_1)

plt.figure(dpi=600)
sim.plot2D()
plt.show()
```
